## Visualize Annual Cattle Sales (Head) Using USDA NASS Data

This notebook creates a time series plot of annual U.S. cattle sales in millions of head, using data from the USDA National Agricultural Statistics Service (NASS).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # figure creation and plot aesthetics
import seaborn as sns  # statistical plotting with built-in themes

sns.set_style('whitegrid')  # clean background with subtle gridlines

cattle_head = pd.read_csv('../data/cattle_sales_head_annual_nass.csv')
```

```python
cattle_head = cattle_head.assign(
    num_head_millions=cattle_head['num_head']/1_000_000
)
```

```python
cattle_head.head()
```

## Line Plot

The chart below shows annual U.S. cattle sales (excluding calves) from USDA NASS, expressed in millions of head.

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(
    data=cattle_head,
    x='year',
    y='num_head_millions',
    marker='s',  # square markers
    linewidth=1.5,
    color='darkblue',
    ax=ax
)

ax.set_title("Cattle Sales Head (Millions)", fontsize=14, weight='bold')
ax.set_xlabel("Year", fontsize=14, weight='bold')
ax.set_ylabel("Number of Head (Millions)", fontsize=14, weight='bold')
ax.tick_params(axis='x', rotation=90, labelsize=14)  # rotate x-axis labels for readability
ax.tick_params(axis='y', labelsize=14)

# Set x-axis to display every year in the dataset
# Get all unique years from the data and sort them
all_years = sorted(cattle_head['year'].unique())
# Set x-tick locations to show every three year
ax.set_xticks(all_years[::3])
# Set x-tick labels to match the years
ax.set_xticklabels(all_years[::3])

fig.tight_layout()
fig.savefig('output/figures/cattle_sales_head_annual.png', dpi=150, bbox_inches='tight')
plt.show()
```
